```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### SELECT APPROPRIATE ROWS

```python
#30/9/2002 is the date where the value of the relative humidity start to collect.
#Hence we need to remove rows before the date 30/9/2002 as it has incomplete data
#Retrieve the last index of 30/9/2002 from the data which is 6551
klangapi[klangapi['Full_Date'] == '1/1/2018'].index[0]
```

```python
#Then remove all of the data from the 1 to 6551 index
#Used .copy() to avoid warnings
klangapiSelRows = klangapi.iloc[133973:].copy().reset_index(drop=True)
klangapiSelRows.head()
```

### DROP UNECESSARY COLUMNS AND CHANGE COLUMN NAMES

```python
klangapiSelRows.drop(columns=['UVB_hourly_avg','API_Main','Full_Date','Month','Years','Site_Id','Site_Location','Time'], inplace = True)
klangapiSelRows.columns=['date','wind_speed','wind_direction','temperature','relative_humidity','nox','no','so2','no2','o3','co','pm10','pm2.5']

klangapiSelRows.shape
```

```python
klangapiSelRows.info()
```

### DEALING WITH MISSING DATA AND INCORRECT DATA TYPE

```python
klangapiSelRows[(klangapiSelRows == 0.0).any(axis=1)]
```

```python
klangapiSelRows[(klangapiSelRows.isnull()).any(axis=1)]
```

```python
klangapiSelRows.info()
```

```python
#Change the incorrect datatype to the correct datatype
#object to float
klangapiSelRows["relative_humidity"] = pd.to_numeric(klangapiSelRows.relative_humidity, errors='coerce')
klangapiSelRows["nox"] = pd.to_numeric(klangapiSelRows.nox, errors='coerce')
klangapiSelRows["no"] = pd.to_numeric(klangapiSelRows.no, errors='coerce')
klangapiSelRows["so2"] = pd.to_numeric(klangapiSelRows.so2, errors='coerce')
klangapiSelRows["no2"] = pd.to_numeric(klangapiSelRows.no2, errors='coerce')
klangapiSelRows["o3"] = pd.to_numeric(klangapiSelRows.o3, errors='coerce')
klangapiSelRows["co"] = pd.to_numeric(klangapiSelRows.co, errors='coerce')
klangapiSelRows["pm10"] = pd.to_numeric(klangapiSelRows.pm10, errors='coerce')
```

```python
#Check data(datatype)
klangapiSelRows.info()
```

```python
#Check missing data
klangapiSelRows.isnull().sum()
```

```python
#fill in missing data with linear interpolation 
klangapiSelRows.interpolate(method ='linear', limit_direction ='both', inplace=True)
```

```python
#Check missing data again
klangapiSelRows.isnull().sum()
```

```python
klangapiSelRows.info()
```

## RESAMPLE THE DATA TO DAILY MEAN AND MAX

```python
#change the date datatype from int to date type
klangapiSelRows['date'] = pd.to_datetime(klangapiSelRows['date'], format='%Y%m%d')
```

```python
#set the date to index
klangapiSelRows.set_index('date', inplace=True)
klangapiSelRows.head()
```

```python
# klangapiHourly = klangapiSelRows.reset_index()
# klangapiHourly.to_csv('klangapiCleanedHourlyPm2.5.csv', encoding='utf-8',index=False)
```

```python
#resample data
klangapiDailyMean = klangapiSelRows.resample('D').mean()
klangapiDailyMean.shape
```

```python
# klangapiDailyMax = klangapiSelRows.resample('D').max()
# klangapiDailyMax.shape
```

```python
#round off data sebb bnyk sangat point kt belakang bila averaging data jadi daily
klangapiDailyMean=klangapiDailyMean.round(4)
klangapiDailyMean.head()
```

```python
klangapiDailyMean.isnull().sum()
```

```python
#262 missing data sebb DOE stop collected data from this period so that why missing
klangapiDailyMean[klangapiDailyMean['wind_direction'].isnull()]
```

```python
#drop that missing data
klangapiDailyMean.dropna(inplace=True)
klangapiDailyMean.shape
```

```python
klangapiDailyMean[(klangapiDailyMean == 0.0).any(axis=1)]
```

```python
#2002-01-01 to 2020-12-31 = 6940 days
#2002-01-01 to 2002-09-30 = 273 days
#273 days removed
#6940 - 273 = 6667 days
#So data train mesti ade 6667 rows
#tapi data from 14/4/2017 - 31/12/2017 takde, so 6667 -262 = 6405 rows
```
