```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read csv, print 5 rows
klang2021 = pd.read_csv('Klang 2021.csv',encoding='latin1')
klang2021.head()
```

```python
#read csv, print 5 rows
klangapi = pd.read_csv('klangapiraw.csv',encoding='latin1')
klangapi.head()
```

```python
klangapi.shape
```

### SELECT APPROPRIATE ROWS

```python
#Then remove all of the data from the 1 to 6551 index
#Used .copy() to avoid warnings
klangapiSelRows = klangapi.iloc[6552:].copy().reset_index(drop=True)
klangapiSelRows.head()
```

### DROP UNECESSARY COLUMNS AND CHANGE COLUMN NAMES

```python
klangapiSelRows.drop(columns=['UVB_hourly_avg','PM2.5 1H (µg/m3)','API_Main','Full_Date','Month','Years','Site_Id','Site_Location','Time'],inplace=True)
klangapiSelRows.columns=['date','wind_speed','wind_direction','temperature','relative_humidity','nox','no','so2','no2','o3','co','pm10']
klangapiSelRows.drop(columns=['nox','no'],inplace=True)

klangapiSelRows.shape
```

```python
klang2021.drop(columns=['STATION ID','LOCATION','API','PM2.5 1H (µg/m3)','SOLAR RADIATION 1H (W/m2)'],inplace=True)
klang2021.columns=['date','pm10','so2','no2','o3','co','wind_direction','wind_speed','relative_humidity','temperature']

klang2021.shape
```

```python
klangapiSelRows.info()
```

```python
klang2021.info()
```

```python
#change the date datatype from int to date type
klang2021['date'] = pd.to_datetime(klang2021['date'], format='%d/%m/%Y %H:%M')

#set the date to index
klang2021.set_index('date', inplace=True)
klang2021.head()
```

```python
#change the date datatype from int to date type
klangapiSelRows['date'] = pd.to_datetime(klangapiSelRows['date'], format='%Y%m%d')

#set the date to index
klangapiSelRows.set_index('date', inplace=True)
klangapiSelRows.head()
```

```python
klangapinews = pd.concat([klangapiSelRows, klang2021], axis=0)
```

```python
klangapinews
```

### DEALING WITH MISSING DATA AND INCORRECT DATA TYPE

```python
klangapinews[(klangapinews == 0.0).any(axis=1)]
```

```python
klangapinews[(klangapinews.isnull()).any(axis=1)]
```

```python
klangapinews.info()
```

```python
#Change the incorrect datatype to the correct datatype
#object to float
klangapinews["relative_humidity"] = pd.to_numeric(klangapinews.relative_humidity, errors='coerce')
klangapinews["so2"] = pd.to_numeric(klangapinews.so2, errors='coerce')
klangapinews["no2"] = pd.to_numeric(klangapinews.no2, errors='coerce')
klangapinews["o3"] = pd.to_numeric(klangapinews.o3, errors='coerce')
klangapinews["co"] = pd.to_numeric(klangapinews.co, errors='coerce')
klangapinews["pm10"] = pd.to_numeric(klangapinews.pm10, errors='coerce')
```

```python
#Check data(datatype)
klangapinews.info()
```

```python
#Check missing data
#rh ade yang 0, logically this value xleh ade 0 so kita kne jadikan nilai 0 tu nan and replace with linear interpolate
klangapinews['relative_humidity']=klangapinews['relative_humidity'].replace(0.0000,np.NaN)
klangapinews.isnull().sum()
```

```python
#fill in missing data with linear interpolation 
klangapinews.interpolate(method ='linear', limit_direction ='forward', inplace=True)
```

```python
#Check missing data again
klangapinews.isnull().sum()
```

```python
klangapinews.info()
```

```python
klangapinews[(klangapinews == 0.0).any(axis=1)]
```

## RESAMPLE THE DATA TO DAILY MEAN AND MAX

```python
#resample data
klangapiDailyMean = klangapinews.resample('D').mean()
klangapiDailyMean.shape
```

```python
klangapiDailyMean[(klangapiDailyMean == 0.0).any(axis=1)]
```

```python
# klangapiDailyMax = klangapiSelRows.resample('D').max()
# klangapiDailyMax.shape
```

```python
#round off data sebb bnyk sangat point kt belakang bila averaging data jadi daily
klangapiDailyMean=klangapiDailyMean.round(4)
klangapiDailyMean.head()
```

```python
klangapiDailyMean.isnull().sum()
```

```python
#262 missing data sebb DOE stop collected data from this 14/4/2017 - 31/12/2017 period so that why missing
klangapiDailyMean[klangapiDailyMean['wind_direction'].isnull()]
```

```python
#drop that missing data
klangapiDailyMean.dropna(inplace=True)
klangapiDailyMean.shape
```

```python
klangapiDailyMean[(klangapiDailyMean == 0.0).any(axis=1)]
```
