### SELECT APPROPRIATE ROWS

```python
#30/9/2002 is the date where the value of the relative humidity start to collect.
#Hence we need to remove rows before the date 30/9/2002 as it has incomplete data
#Retrieve the last index of 30/9/2002 from the data which is 6551
klangapi[klangapi['Full_Date'] == '30/9/2002'].index[-1]
```

```python
klangapiSelRows.drop(columns=['UVB_hourly_avg','PM2.5 1H (µg/m3)',
                              'API_Main','Full_Date',
                              'Month','Years','Site_Id',
                              'Site_Location','Time'],inplace=True)

klangapiSelRows.columns=['date','wind_speed','wind_direction',
                         'temperature','relative_humidity',
                         'nox','no','so2','no2','o3','co','pm10']

klangapiSelRows.shape
```

## RESAMPLE THE DATA TO DAILY MEAN AND MAX

```python
#change the date datatype from int to date type
klangapiSelRows['date'] = pd.to_datetime(klangapiSelRows['date'], format='%Y%m%d')
```

```python
#set the date to index
klangapiSelRows.set_index('date', inplace=True)
klangapiSelRows.head()
```

```python
#resample data to daily mean , max
klangapiDailyMean = klangapiSelRows.resample('D').mean()
klangapiDailyMax = klangapiSelRows.resample('D').max()
klangapiDailyMean.shape
```

```python
# klangapiDailyMax = klangapiSelRows.resample('D').max()
# klangapiDailyMax.shape
```

```python
#round off data sebb bnyk sangat point kt belakang bila averaging data jadi daily
klangapiDailyMean=klangapiDailyMean.round(4)
klangapiDailyMean.head()
```

```python
klangapiDailyMean.isnull().sum()
```

```python
#262 missing data sebb DOE stop collected data from this 14/4/2017 - 31/12/2017 period so that why missing
klangapiDailyMean[klangapiDailyMean['wind_direction'].isnull()]
```

```python
#drop that missing data
klangapiDailyMean.dropna(inplace=True)
klangapiDailyMean.shape
```

```python
klangapiDailyMean[(klangapiDailyMean == 0.0).any(axis=1)]
```

```python
#2002-01-01 to 2020-12-31 = 6940 days
#2002-01-01 to 2002-09-30 = 273 days
#273 days removed
#6940 - 273 = 6667 days
#So data train mesti ade 6667 rows
#tapi data from 14/4/2017 - 31/12/2017 takde, so 6667 -262 = 6405 rows
```

```python
klangapiDailyMean.reset_index(inplace=True)
klangapiDailyMean.head()
```
