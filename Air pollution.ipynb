```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read csv, print 5 rows
klangapi = pd.read_csv('klangapi.csv')
klangapi.head()
```

```python
#change column names for simplicity
klangapi.columns=['date', 'month', 'wind_speed', 'wind_direction', 'temperature','uvb', 'relative_humidity', 'nox', 'no','so2', 'so2_day1', 'so2_day2','so2_day3', 'no2','no2_day1', 'no2_day2', 'no2_day3', 'o3', 'o3_day1','o3_day2', 'o3_day3','co','co_day1','co_day2','co_day3','pm10','pm10_day1','pm10_day2','pm10_day3']
```

```python
#check variables
klangapi.info()
```

```python
#check is missing data exist
klangapi.isnull().sum()
```

```python
#check how many unique data in each variables
klangapi.nunique()
```

```python
#check duplicate data
klangapi[klangapi.duplicated() == True]
```

```python
#extract several data and put in new dataframe

klangapinew = klangapi.loc[:,['date', 'month', 'wind_speed', 'wind_direction', 'temperature','uvb', 'relative_humidity', 'nox', 'no', 'so2', 'no2', 'o3', 'co', 'pm10']]
klangapinew.head()
```

```python
#check if inconsistent data exist such as value 0
klangapinew[(klangapinew == 0.0).any(axis=1)]
```

```python
#aggregaate data according to month with mean
data = klangapinew.groupby('month', as_index=False).agg({'wind_speed':'mean', 'wind_direction':'mean', 'temperature':'mean', 'uvb':'mean', 'relative_humidity':'mean', 'nox':'mean', 'no':'mean', 'so2':'mean', 'no2':'mean', 'o3':'mean', 'co':'mean', 'pm10':'mean'})
data
```

```python
#sort the data type based on month
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
data['month'] = pd.Categorical(data['month'], categories=month_order, ordered=True)
data.sort_values(by=['month'], inplace=True)
data
```

```python
data.info()
```

```python
# data.plot(x='month', y=['pm10'])
plt.figure(figsize=(15, 5))
plt.plot(data['month'],data['pm10'])
plt.show()
```

```python
# data.plot(x='month', y=['co'])
plt.figure(figsize=(15, 5))
plt.plot(data['month'],data['co'])
plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.plot(data['month'],data['so2'])

plt.show()
```

```python
#data.plot(x='month', y=['no','o3','no2','so2','no','nox'])
plt.figure(figsize=(15, 5))

plt.plot(data['month'],data[['no','o3','no2','no','nox']])
plt.legend(['no','o3','no2','no','nox'])

plt.show()
```
